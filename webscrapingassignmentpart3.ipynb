8. Write a python program to extract information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day
extended forecast display for the city. The data should include period, short description, temperature and
description. 

```python
from bs4 import BeautifulSoup
import requests
```

```python
page1=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.X8z0mNgzbIU")
```

```python
page1.content
```

```python
soup=BeautifulSoup(page1.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
Day=[]
desc=[]
temp=[]
```

```python
day=soup.find_all('p',class_="period-name")
day[0:]
```

```python
for i in day:
    Day.append(i.text)
Day[0:]
```

```python
des=soup.find_all('p',class_="short-desc")
des[0:]
```

```python
for i in des:
    desc.append(i.text)
desc[0:]
```

```python
t=soup.find_all('p',class_=["temp temp-low","temp temp-high"])
t[0:]
```

```python
for i in t:
    temp.append(i.text)
temp[0:]
```

```python
print(len(Day),len(desc),len(temp))
```

```python
import pandas as pd
weather=pd.DataFrame({})
weather['Day']=Day[:9]
weather['Temperature']=temp[:9]
weather['Description']=desc[:9]
```

```python
weather
```

7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. 

```python
page2=requests.get("https://www.amazon.in/s?i=electronics&bbn=1389401031&rh=n%3A976419031%2Cn%3A1389401031%2Cn%3A1389432031%2Cp_36%3A-2000000&dc&qid=1607273906&rnid=1389401031&ref=sr_nr_n_3")
```

```python
page2.content
```

```python
soup=BeautifulSoup(page2.content,'html.parser')
print(soup.prettify())
```

```python
ProductName=[] 
Price=[]
ImageURL=[]  
AverageRating=[]
```

```python
titles=soup.find_all('a',class_="a-link-normal a-text-normal")
titles[0:]
```

```python
for i in titles:
    full_title=""
    for j in i.find_all('span'):
        full_title=full_title+" "+j.get_text()
    print(full_title)
    ProductName.append(full_title)
```

```python
price=soup.find_all('span',class_="a-price-whole")
price[0:]
```

```python
for i in price:
    Price.append(i.get_text())
Price[0:]
```

```python
img=soup.find_all('img',class_="s-image")
img[0:]
```

```python
for i in img:
    ImageURL.append(i.get_text())
ImageURL[0:]
```

```python
rate=soup.find_all('a',class_="a-popover-trigger a-declarative")
rate[0:]
```

```python
for i in rate:
    full_rate=""
    for j in i.find_all('span'):
        full_rate=full_rate+" "+j.get_text()
    print(full_rate)
    AverageRating.append(full_rate)
```

```python
print(len(ProductName),len(Price),len(AverageRating))
```

```python
import pandas as pd
mobiles=pd.DataFrame({})
mobiles['ProductName']=ProductName[:24]
mobiles['Price']=Price[:24]
mobiles['AverageRating']=AverageRating[:24]
```
