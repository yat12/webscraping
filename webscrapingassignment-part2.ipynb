4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
page4=requests.get("https://bookpage.com/reviews")
```

```python
page4.content
```

```python
soup=BeautifulSoup(page4.content,'html.parser')
print(soup.prettify())
```

```python
Name=[]
Author=[]
Genre=[]
Review=[]
```

```python
titles=soup.find_all('h4',class_="italic")
titles[0:5]
```

```python
for i in titles:
    full_title=""
    for j in i.find_all('a'):
        full_title=full_title+" "+j.get_text()
    print(full_title)
    Name.append(full_title)
```

```python
author=soup.find_all('p',class_="sans bold")
author[0:5]
```

```python
for i in author:
    Author.append(i.text.replace('\n',''))
Author[0:5]
```

```python
genr=soup.find_all('p',class_="genre-links hidden-phone")
genr[0:5]
```

```python
for i in genr:
    full_title=""
    for j in i.find_all('a'):
        full_title=j.get_text()
    print(full_title)
    Genre.append(full_title)
```

```python
review=soup.find_all('p',class_="excerpt")
review[0:5]
```

```python
for i in review:
    Review.append(i.text.replace('\n',''))
Review[0:5]
```

```python
print(len(Name),len(Author),len(Review),len(Genre))
```

```python
import pandas as pd
books=pd.DataFrame({})
books['Book Name']=Name[:5]
books['Author']=Author[:5]
books['Genre']=Genre[:5]
books['Review']=Review[:5]
```
